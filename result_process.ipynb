```python
# import data
from google.colab import files
uploaded = files.upload()

import json
data = next(iter(uploaded.values()))
d = json.loads(data.decode())
print(d)
```

```python
# 1200 rows
import pandas as pd
dataArr = []

for line in d:
  df_res_flattened = pd.json_normalize(line, record_path =['result'])
  for res in df_res_flattened.iterrows():
    dataLine = []
    dataLine.append(line['queryId'])
    dataLine.append(line['scenarioId'])
    dataLine.append(line['features']['disability'])
    dataLine.append(line['features']['race'])
    dataLine.append(line['features']['religion'])
    dataLine.append(line['features']['gender'])
    for i in range(0, 9):
      dataLine.append(res[1][i])
    dataArr.append(dataLine)

df = pd.DataFrame(dataArr, columns=["queryId", "scenarioId", "disability", "race", "religion", "gender", "happiness", "anger", "fear", "sadness", "disgust", "surprise", "exciting", "pleasant", "powerful"])
print(df)
```
